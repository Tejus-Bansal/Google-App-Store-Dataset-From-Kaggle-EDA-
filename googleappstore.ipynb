## Deleting duplicate rows from the dataset

```python
df_no_duplicate = df.drop_duplicates(keep="first")
```

```python
df_no_duplicate
```

```python
df_no_duplicate.info()
```

## Filling the null values in the "Rating" column with the mean of the entire 'Rating' column

```python
df_no_duplicate.loc[:, "Rating"] = df_no_duplicate["Rating"].fillna(df_no_duplicate["Rating"].mean())
```

```python
df_no_duplicate.info()
```

```python
df_no_duplicate.columns
```

## Removing the rest of the rows with null values

```python
df_no_duplicate = df_no_duplicate.dropna(axis=0)
```

```python
df_no_duplicate.info()
```

```python
df_no_duplicate.head(15)
```

## Pre-Processing the 'Size' column
#### ---> replacing 'M' with '000' and 'k' with ''

```python
df_no_duplicate['Size'].unique()
```

```python
df_no_duplicate.loc[:,'Size'] = df_no_duplicate['Size'].str.replace('k','')
```

```python
df_no_duplicate.loc[:,'Size'] = df_no_duplicate['Size'].str.replace('M','000')
```

#### filling the cell containing 'Varies with device' value with the mean of the entire 'Size' column

```python
df_no_duplicate.loc[:,'Size'] = df_no_duplicate['Size'].replace('Varies with device', np.nan)
```

```python
df_no_duplicate.loc[:,'Size']=df_no_duplicate['Size'].astype(float)
```

```python
df_no_duplicate.loc[:,'Size']=df_no_duplicate['Size'].fillna(df_no_duplicate['Size'].mean())
```

```python
df_no_duplicate['Size'].unique()
```

```python
df_no_duplicate.loc[:,'Installs'] = df_no_duplicate['Installs'].str.replace(',', '').str.replace('+', '')
```
